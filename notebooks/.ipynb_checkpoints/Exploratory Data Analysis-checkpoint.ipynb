**The dataset used for this thesis originates from Kaggle. The dataset contains various variables, including lifestyle and medical data. The dataset consist of 35 features, and 2149 entries. The target variable is the Diagnosis variable, which is a binary variable where 1 refers to the patient having Alzheimer's Disease and 0 referring to the patient not having Alzheimer's disease. Further information and the EDA will be explained below.**

# Data Description

**In this Exploratory Data Analysis (EDA) several things have become clear concerning the structure and completeness of the data. First of all, the EDA points out that the overall data quality is quite high. The dataset consist of 35 variables and 2149 entries.**

Furthermore, the corpus level analysis points out that there are no null-values found in any of the columns. Furthermore, potential duplicates in PatientID have been checked, as this would be disrupting for the rest of the research. No duplicate PatientID's have been found, which is a positive. The datatypes have been checked, and boxplots have been generated to detect potential outliers. One noticable finding, is that the final column, namely DoctorInCharge, is blanked out due to privacy reasons. Therefore, the column DoctorInCarge has been removed from the dataset.

**Univariate Analysis**


The Univariate Analysis shows an interesting finding. Quite some of the variables are measured on a ordinal bounded scale (0-10 e.g.). This is not something to be resolved in the EDA. However, this is an important finding for the model development stage of this research, as this may affect model performance and behavior. Furthermore the distribution of age has been checked, as this might be an important variable for the Alzheimer's Disease detection. The distribution is listed below:

![image](../images/Age_distribution.png)

As visible in the graph above, the age is quite evenly distributed and ranges from 60-90. Considering Alzheimer's Disease typically develops mid-60s or later, this range makes sense. 

**Numerical Variables and Diagnosis**

Furthermore, the target variable has been invested, namely the Diagnosis variable, which has a proportion of 35% are not diagnosed with Alzheimer's and 65% are diagnosed with Alzheimer's. Additionally, the feature MMSE has been plotted next to the Diagnosis feature, as MMSE is assumed to be highly predictive for the target variable. The results are displayed below:

![image](../images/MMSE_diagnosis.png)

A clear difference is observed in MMSE scores between diagnosed and non-diagnosed patients. Individuals diagnosed with Alzheimer’s disease show notably lower MMSE scores compared to those without a diagnosis. This indicates that MMSE is strongly associated with cognitive impairment and serves as a key predictor for Alzheimer’s disease.

The same boxplots have been generated for the variables BMI, Age, and PhysicalActivity. The results of these are that the boxplots are very similar, meaning they are individually weak predictors. However, the combination of several features might increase predictive performance, and therefore they are still considered in further research. 

**Medical and Lifestyle variables**


Additionally, more boxplots where generated for both lifestyle and medical variables seperately. In addition to MMSE, functional assessment and activities of daily living (ADL) scores show a clear distinction between diagnosed and non-diagnosed patients. Individuals with Alzheimer’s disease exhibit notably lower scores on both measures, indicating reduced functional and daily living capabilities. The distributions of these variables show a clear shift between the two groups, with less overlap compared to lifestyle variables. This suggests that functional and cognitive measures have strong discriminative power and are likely to be key predictors in the modeling phase. The corresponding boxplots are displayed below: 

![image](../images/ADL_diagnosis.png)
![image](../images/FA_diagnosis.png)

**Categorical Variables Analysis**


Finally, the analysis of categorical variables shows only minor differences in Alzheimer’s diagnosis rates across groups. It is presented below: 

![image](../images/crosstab.png)

Variables such as diabetes, smoking, and family history show similar proportions of diagnosed people. This indicates they are weak predictors for the target variable. From the four variables plotted, the hypertension variable is the most telling one, but still the impact seems to remain limited. Overall, the categorical medical and lifestyle variables have lower discriminative power compared to the cognitive and functional variables. ]


**Key EDA Findings and Implications**

Overall, the EDA has revealed that cognitive and functional variables, and more particularly MMSE, Functional Assesssment, and ADL, have strong predictive potential for the Alzheimer's diagnosis prediction. Compared to the functional and cognitive variables, the lifestyle and demographic variables seem to have limited individual predictive power. These findings suggest that a model trained on cognitive and functional variables is likely to outperform a model trained solely on lifestyle features. However, an important nuance in this context is that the EDA only reveals individual predictive power to the target variable, and the combination of variables might be a stronger predictor. However, this is something to investigate further in the modelling phase, but is important to note.

```python
#Plot distributions
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()

sns.histplot(df['BMI'], kde=True)
plt.title("BMI Distribution")
plt.show()

sns.histplot(df['MMSE'], kde=True)
plt.title("MMSE Distribution")
plt.show()
```

```python
#plot MMSE vs Diagnosis as this might be a relevant variable for Diagnosis prediction 
sns.boxplot(x='Diagnosis', y='MMSE', data=df)
```

```python
#check some other variables for potential predictive strength
sns.boxplot(x='Diagnosis', y='Age', data=df)
plt.show()

sns.boxplot(x='Diagnosis', y='BMI', data=df)
plt.show()

sns.boxplot(x='Diagnosis', y='PhysicalActivity', data=df)
plt.show()
```

```python
#Check some of the medical variables 
# Blood pressure
sns.boxplot(x='Diagnosis', y='SystolicBP', data=df)
plt.title("Systolic Blood Pressure vs Diagnosis")
plt.show()

sns.boxplot(x='Diagnosis', y='DiastolicBP', data=df)
plt.title("Diastolic Blood Pressure vs Diagnosis")
plt.show()

```

```python
#some more medical variables 
# Cholesterol
sns.boxplot(x='Diagnosis', y='CholesterolTotal', data=df)
plt.title("Total Cholesterol vs Diagnosis")
plt.show()

sns.boxplot(x='Diagnosis', y='CholesterolLDL', data=df)
plt.title("LDL Cholesterol vs Diagnosis")
plt.show()

sns.boxplot(x='Diagnosis', y='CholesterolHDL', data=df)
plt.title("HDL Cholesterol vs Diagnosis")
plt.show()
```

```python
#categorical variables vs diagnosis

print("Family History")
print(pd.crosstab(df['FamilyHistoryAlzheimers'], df['Diagnosis'], normalize='index'))

print("\nDiabetes")
print(pd.crosstab(df['Diabetes'], df['Diagnosis'], normalize='index'))

print("\nHypertension")
print(pd.crosstab(df['Hypertension'], df['Diagnosis'], normalize='index'))

print("\nSmoking")
print(pd.crosstab(df['Smoking'], df['Diagnosis'], normalize='index'))
```
